```python
# Import yolo from ultralytics
from ultralytics import YOLO
```

**ultralytics has recently released YOLOv8 model in yolo family, it has five versions**

1. YOLOv8 Nano: [yolov8n.pt]

   Lightweight version optimized for resource-constrained environments with lower model size and faster inference.more suitable    for low configurations systems.
<br>

2. YOLOv8 Small:[yolov8s.pt]

   Compact model balancing speed and accuracy suitable for real-time object detection applications. this is also optimized for      low configurations systems
<br>

3. YOLOv8 Medium:[yolov8m.pt]

   Moderate-sized model offering a good trade-off between accuracy and inference speed for various applications. 
<br>

4. YOLOv8 Large: [yolov8l.pt]

   Larger model designed to achieve higher accuracy at the cost of slightly slower inference, suitable for more demanding tasks.
<br>

5. YOLOv8 Extra Large: [yolov8x.pt]

   High-capacity model with increased parameters, providing state-of-the-art accuracy for complex object detection scenarios.

```python
# Preparing yolo Model [Iam using medium model so that i can get moderate results]
model = YOLO('yolov8m.pt')
```

**In yolov8 model we have three modes and three tasks**

1. train

2. val

3. predict


**Tasks**

1. Detect : For object detection

2. Segment : For Image segmentation

3. classify : For Image classification

<br>
Now we are going to use only Detect task in this project

```python
# Now we are using first mode that is TRAIN
# Here I reduced the image size to 608 to avoid more computing calculations, default image size is 640
### Training the Our yolo Model
model.train(batch = 8,imgsz=608, data = 'data.yaml',epochs = 5,device='cpu')
```

**After training results will be saved in runs folder.**


**Trained images,confusion matrix,weights[ model ],etc. These things will be saved inside the runs forlder**

```python
# Now validate our model:
!yolo task=detect mode=val model=runs\detect\train2\weights\best.pt data=data.yaml
```
